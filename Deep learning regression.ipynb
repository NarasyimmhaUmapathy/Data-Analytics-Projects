```python
# Model creation and training

X = df.drop("price",axis=1).values #TF cant work with dataframes, has to be np array

y = df["price"].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
 X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df.head()
```

```python
model = Sequential()

model.add(Dense(19,activation="relu")) #num of neurons corresponds to num of features
model.add(Dense(19,activation="relu"))
model.add(Dense(19,activation="relu"))
model.add(Dense(19,activation="relu"))

model.add(Dense(1)) #last layer neuros corresp.s to number of labels 

model.compile(optimizer="adam",loss="mse") #optimizes gradient descent function
```

```python
model.fit(x=X_train,y=y_train,
          validation_data=(X_test,y_test),
         batch_size=(128),epochs=400)
```
